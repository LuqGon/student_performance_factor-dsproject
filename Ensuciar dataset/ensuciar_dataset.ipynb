```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("StudentPerformanceFactors_original.csv")
print(df.shape)
df.head()
```

```python
df.loc[df.sample(frac=0.01).index, "Attendance"] = np.nan
```

```python
def ensuciar_categoria(valor):
    if valor == "Low":
        return np.random.choice(["low", "Low ", "l0w", "LOW", "Lo w"])
    if valor == "Medium":
        return np.random.choice(["med", "Medium ", "MEDIUM", "Mediu m", "m3dium"])
    if valor == "High":
        return np.random.choice(["hig", "High ", "HIGH", "H1gh", "Hi gh"])
    return valor

df["Teacher_Quality"] = df["Teacher_Quality"].apply(ensuciar_categoria)
```

```python
duplicados = df.sample(frac=0.2)
df = pd.concat([df, duplicados]).sample(frac=1).reset_index(drop=True)
```

```python
df.to_csv("StudentPerformanceFactors.csv", index=False)
```
